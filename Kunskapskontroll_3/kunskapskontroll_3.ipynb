```python
## Ny dataset, ta bort rader som inte innehåller värde "total_bedrooms"
df_housing = df.dropna()
print(df_housing.isnull().sum())
df_housing.info()
```

```python
## Ny data set
df_housing = df_housing[(df_housing["median_house_value"] <=  500000 ) & (df_housing["housing_median_age"] != 52.0)]
df_housing.info()
```

```python
# Sammanfatta statistik över huset förhållande till havet
# 
grp_ocean_proximity = df_housing["ocean_proximity"].unique()
grp_ocean_proximity_value = df_housing["ocean_proximity"].value_counts()
#print(grp_ocean_proximity)
print(grp_ocean_proximity_value)

plt.figure(figsize=(15,5))
plt.bar(grp_ocean_proximity_value.index,grp_ocean_proximity_value.values,color="lightgreen",edgecolor="black")
plt.xlabel("Category")
plt.ylabel("Quantity")
plt.title("Ocean_proximity")

plt.show()
```

```python
# Sammanfatta statistik, över median inkomst vs huset värde (samma graf som tididgare med rensad data)
plt.scatter(df_housing['median_income'],df_housing['median_house_value'], alpha=0.7,edgecolor='k')
plt.title('Median House vs Median Incomst',fontsize=14)
plt.xlabel('Median Incomst',fontsize=12)
plt.ylabel('Median House Value',fontsize=12)
plt.grid(True,alpha=0.6)

## Nästan liknade graf som tididgare test tak värde är borta,
## Max inkomsten är forfrande runt ~50K.
```

```python
# Sammanfatta statistik, över media inkomst
income = df_housing['median_income']

avg_median_income = income.median()
print(avg_median_income)


# Histogram över media inkomst
plt.hist(income, bins=30, alpha=0.5)

plt.xlabel('Median Income')
plt.ylabel('Count')
plt.title('Distribution of Median Incomes')


plt.show()

## förklaring - Detta är en positivt snedfördelning, 
## medelvärdet är större än medianen och läget är det minsta värdet i datamängden.
```

```python
# Sammanfatta statistik, över latitut och longtitut

# Create figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

#  scatter plot - new dataset
axes[0].scatter(df_housing['longitude'], df_housing['latitude'], s=1)
axes[0].set_title("Latitude vs. Longitude")
axes[0].set_xlabel("Longitude")
axes[0].set_ylabel("Latitude")


#  scatter plot - Old dataset
axes[1].scatter(df['longitude'], df['latitude'], s=1)
axes[1].set_title("Latitude vs. Longitude")
axes[1].set_xlabel("Longitude")
axes[1].set_ylabel("Latitude")

plt.show()

## Presentera båda dataset, den fullstände och den nya med rensning av max data.
## Båda graferna ser nästan exat likadan ut :-|
```
